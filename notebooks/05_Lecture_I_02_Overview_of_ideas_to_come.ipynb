# An overview of key ideas

The two key ideas that we develop in this notebook is that of matrices that are invertible and those that are not. We will not formalize these concepts, but will build the intuition for them. We will see that the first matrix, $A$, is invertible and the second one, $M$, is not. We will also see that the columns of $A$ are independent and span $\mathbb{R}^{3}$, while the columns of $M$ are not independent and do not span $\mathbb{R}^{3}$.

## Moving from vectors to matrices

Consider two position vectors $\mathbf{u} , \mathbf{v} \in \mathbb{R}^{3}$.  They can be expressed as column vectors.

$$ \mathbf{u}=\begin{bmatrix}1\\-1\\0\end{bmatrix}, \; \mathbf{v}=\begin{bmatrix}0\\1\\-1\end{bmatrix} \tag{1} $$

We can add constant scalar multiples of these vectors, with the scalars $x_1$ and $x_2$.

$$ {x}_{1}\mathbf{u}+{x}_{2}\mathbf{v}=\mathbf{b}, \text{ where }\mathbf{b} \in \mathbb{R}^{2} \tag{2} $$

This is simple scalar-vector multiplication and vector-vector addition.  It should be clear that if we combine all possible scalar multiples of these vectors, that we start filling a plane through the origin.  Adding a third vector that is not in this plane will, extends this idea to fill all of three-dimensional space.

$$ \mathbf{w}=\begin{bmatrix}0\\0\\1\end{bmatrix} \tag{3} $$

We now have (4) below.

$$
\begin{align*}
&{x}_{1}\mathbf{u}+{x}_{2}\mathbf{v}+{x}_{3}\mathbf{w}=\mathbf{b}, \text{ where }\mathbf{b} \in \mathbb{R}^{3} \\ \\
\end{align*}
\tag{4}
$$

The solution vector $\mathbf{x}$, shown in (5).

$$ \mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}\tag{5}$$

Let us now write (4) is the form $A \mathbf{x} = \mathbf{b}$. We want to find the vector $\mathbf{x}$ that satisfies (4).

$$
\begin{align*}
&\text{Matrix view: } &\begin{bmatrix} 1 & 0 & 0 \\ -1 & 1 & 0 \\ 0 & -1 & 1 \end{bmatrix}\begin{bmatrix} { x }_{ 1 } \\ { x }_{ 2 } \\ { x }_{ 3 } \end{bmatrix} = \begin{bmatrix} { x }_{ 1 } \\ -{ x }_{ 1 }+{ x }_{ 2 } \\ -{ x }_{ 2 }+{ x }_{ 3 } \end{bmatrix}
\end{align*}
\tag{6}
$$

We create the matrix of coefficients, $A$, in code below.

```python
# Create the matrix of coefficients and assign it to the variable `A`
A = Matrix([[1, 0, 0], [-1, 1, 0], [0, -1, 1]]) # Rows as lists of a list
# Return the matrix `A`
A
```

There is also the column-view of matrix-vector multiplication as opposed to the row view.  Matrices are seen as column, representing vectors.  Each element of the column vector $\mathbf{x}$ is a scalar multiple of the corresponding column in the matrix $A$.

$$
\begin{align*}
&\text{Column view: } \, { x }_{ 1 }\begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix}+{ x }_{ 2 }\begin{bmatrix} 0 \\ 1 \\ -1 \end{bmatrix}+{ x }_{ 3 }\begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix} = \mathbf{b} = \begin{bmatrix} { x }_{ 1 } \\ -{ x }_{ 1 }+{ x }_{ 2 } \\ -{ x }_{ 2 }+{ x }_{ 3 } \end{bmatrix} = {x}_{1}\mathbf{u}+{x}_{2}\mathbf{v}+{x}_{3}\mathbf{w}
\end{align*}
\tag{7}
$$

If we consider the vector, $\mathbf{b}$, in $A \mathbf{x} = \mathbf{b}$, then we can write (8) from (6).

$$ \begin{bmatrix} { x }_{ 1 } \\ -{ x }_{ 1 }+{ x }_{ 2 } \\ -{ x }_{ 2 }+{ x }_{ 3 } \end{bmatrix} = \begin{bmatrix}{b}_{1}\\{b}_{2}\\{b}_{3}\end{bmatrix} \tag{8} $$

By substitution this can be _converted_ to (9).

$$ \begin{bmatrix} { x }_{ 1 } \\ { x }_{ 2 } \\ { x }_{ 3 } \end{bmatrix} = \begin{bmatrix} { b }_{ 1 } \\ { b }_{ 1 }+{ b }_{ 2 } \\ { b }_{ 1 }+{ b }_{ 2 }+{ b }_{ 3 } \end{bmatrix} \tag{9} $$

This is, in fact, a contrived example.  Looking closely, we can write the right-hand side of (9) as a matrix times $\mathbf{b}$.

$$
\mathbf{x} = \begin{bmatrix} { x }_{ 1 } \\ { x }_{ 2 } \\ { x }_{ 3 } \end{bmatrix} = \begin{bmatrix} { b }_{ 1 } \\ { b }_{ 1 }+{ b }_{ 2 } \\ { b }_{ 1 }+{ b }_{ 2 }+{ b }_{ 3 } \end{bmatrix} = \begin{bmatrix}1&0&0\\1&1&0\\1&1&1\end{bmatrix}\begin{bmatrix}{b}_{1}\\{b}_{2}\\{b}_{3}\end{bmatrix}
\tag{10}
$$

This matrix is the inverse of $A$ such that $\mathbf{x} = A^{-1} \mathbf{b}$.  We will learn much more about matrix inverses and whether matrices are invertible.  The `sympy` library has a method called `.inv()` that will calculate the inverse.

```python
# Return the inverse of `A`
A.inv()
```

As an aside, we will come across is _lower triangular form_.  Notice that all the values above the _main diagonal_ (values from the top-left to the bottom-right) as zero value.  This matrix is _lower triangular_.

The standout feature of the example so far is that the vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{w}$, that were used to form the columns of the matrix $A$, are not in the same plane. We have seen that a linear combination of them (adding scalar multiples of $\mathbf{u}, \mathbf{v},$ and $\mathbf{w}$, as in $x_{1}\mathbf{u} + x_{2}\mathbf{v} + x_{3}\mathbf{w}$) can produce any vector, $\mathbf{b}$, in the three-dimensional space. We will learn that this is so, because the matrix of coefficients, $A$, is invertible, and we found $\mathbf{x} = A^{-1}\mathbf{b}$.

Now, let's replace $\mathbf{w}$ such that it is in the same plane as $\mathbf{u}$ and $\mathbf{v}$, and create a matrix $M$, with the last column being our new vector $\mathbf{w}$.  Unlike our original $\mathbf{w}$, this $\mathbf{w}$ is _special_.

```python
M = Matrix([[1, 0, -1], [-1, 1, 0], [0, -1, 1]]) # Creating a matrix and putting
# it into a computer variable called M
M # Displaying it to the screen
```

To illustrate this problem, we cereate some mathematical variables.

```python
# Create the mathematical variables `x1`, `x2`, `x3`, `b1`, `b2`, and `b3` as symbols
x1, x2, x3, b1, b2, b3 = symbols('x1, x2, x3, b1, b2, b3') # Creating algebraic symbols
```

We can now create the solution vector $\mathbf{x}$.

```python
# Create the column vector x with entries `x1`, `x2`, and `x3`
# and assign it to the variable `x_vect`
x_vect = Matrix([[x1], [x2], [x3]]) # Giving this columns vector a computer

# Display the column vector `x_vect`
x_vect
```

Now, $M \mathbf{x}$ can be calculated.

```python
# Calculate the matrix-vector product of `M` and `x_vect`
M @ x_vect
```

For the solution vector, we have three equations that we can create.

$$ { x }_{ 1 }-{ x }_{ 3 }={ b }_{ 1 }\\ { x }_{ 2 }-{ x }_{ 1 }={ b }_{ 2 }\\ { x }_{ 3 }-{ x }_{ 2 }={ b }_{ 3 } \tag{11}$$

Adding the left and righ-hand sides gives us $x_1 - x_3 + x_2 - x_1 + x_3 - x_2 = 0 = b_1 + b_2 + b_3$.  

This leaves us with a constraint for $\mathbf{b}$ such that $ 0={ b }_{ 1 }+{ b }_{ 2 }+{ b }_{ 3 } $.

The problem is clear to conceptualize geometrically as the new $\mathbf{w}$ is in the same plane as $\mathbf{u}$ and $\mathbf{v}$.  In essence $\mathbf{w}$ did not add anything.  All combinations of $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{w}$ will still be in the same plane.

The first $\mathbf{w}$ above created a matrix with three independent columns and their linear combinations could fill all of three-dimensional space (we say they span $\mathbb{R}^{3}$).  This made the first matrix _invertible_ as opposed to the second one, $M$, which is not invertible.

Below, we create the three column vectors and check that $- \mathbf{u} + \left( - \mathbf{v} \right) = \mathbf{w}$.  A linear combination of columns (vectors) one and two, produces three.

```python
# Create the column vector u, v, and w
u = Matrix([[1], [-1], [0]])
v = Matrix([[0], [1], [-1]])
w = Matrix([[-1], [0], [1]])
u, v, w
```

```python
# Return -v
-v
```

```python
# Return -u - v
-u - v
```

```python
# Return w
w
```

```python
# Determine of -u - v equals w
-u - v == w
```

We note that $\mathbf{w}$ is a linear combination of $\mathbf{u}$ and $\mathbf{v}$, specifically $\mathbf{w} = -\mathbf{u} - \mathbf{v}$.
